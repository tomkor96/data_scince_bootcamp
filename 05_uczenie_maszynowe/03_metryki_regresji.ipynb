Interpretacja graficzna

```python
def plot_regression_results(y_true, y_pred): 
    results = pd.DataFrame({'y_true': y_true, 'y_pred': y_pred}) 
    min = results[['y_true', 'y_pred']].min().min() 
    max = results[['y_true', 'y_pred']].max().max() 
    fig = go.Figure(data=[
        go.Scatter(x=results['y_true'], y=results['y_pred'], mode='markers'), 
        go.Scatter(x=[min, max], y=[min, max])], 
                    layout=go.Layout(showlegend=False, width=800, height=500, 
                                     xaxis_title='y_true', 
                                     yaxis_title='y_pred', 
                                     title='Regression results'))
    fig.show() 
    
plot_regression_results(y_true, y_pred)
```

```python

y_true = 100 + 20 * np.random.randn(1000) 
y_pred = y_true + 10 * np.random.randn(1000) 
results = pd.DataFrame({'y_true': y_true, 'y_pred': y_pred}) 
results['error'] = results['y_true'] - results['y_pred'] 
px.histogram(results, x='error', nbins=50, width=800)
```

Średni błąd bezwzględny

```python
# Średni bezwzględny błąd procentowy MAPE (mean absolute percentage error) informuje on o średniej wielkości błędów prognoz dla okresu testowego, wyrażonych w procentach
def mean_absolute_error(y_true, y_pred):
    return abs(y_true - y_pred).sum() / len(y_true)

mean_absolute_error(y_true, y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_true, y_pred)
```

Błąd średniokwadratowy

```python
def mean_squared_error(y_true, y_pred):
    return ((y_true - y_pred) ** 2).sum() / len(y_true)

mean_squared_error(y_true, y_pred)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_true, y_pred)
```

Pierwiastek błędu średniokwadratowego

```python
# jak średnio nasze predykcje odbiegają od wartości prawdziwych
def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(((y_true - y_pred)**2).sum() / len(y_true))

root_mean_squared_error(y_true, y_pred)
```

```python
np.sqrt(mean_squared_error(y_true, y_pred))
```

Błąd maksymalny

```python
# Najgorszy przypadek pomyłki modelu
def max_error(y_true, y_pred):
    return abs(y_true - y_pred).max()
max_error(y_true, y_pred)
```

```python
from sklearn.metrics import max_error

max_error(y_true, y_pred)
```

R2 score - współczynnik determinacji

```python
from sklearn.metrics import r2_score
r2_score(y_true, y_pred) # im bliżej 1, tym lepiej
```
