```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

Tips data

```python
df = sns.load_dataset('tips')
df.head(10)
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T # dodanie danych nieliczbowych (kategorycznych)
```

```python
sns.relplot(data=df, x='total_bill', y='tip')
```

```python
sns.set(font_scale = 1.2)
sns.relplot(data=df, x='total_bill', y='tip', size = 'size', hue='size')
```

```python
sns.relplot(data=df, x='total_bill', y='tip', size = 'size', hue='size', col='time')
```

```python
sns.relplot(data=df, x='total_bill', y='tip', size = 'size', hue='size', col='time', row='smoker')
```

```python
sns.catplot(data=df, x='day', y='total_bill')
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='swarm')
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='box')
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='violin')
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='bar')
```

```python
sns.catplot(data=df, x='day', kind='count')
```

Titanic data

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.catplot(data=df, x='deck', kind='count', palette='Blues')
```

```python
sns.catplot(data=df, y='deck', hue='deck', legend=False, kind='count', palette='Blues')
```
