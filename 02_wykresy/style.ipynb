```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
plt.plot(x, x_sin)
```

```python
plt.style.use('ggplot')
plt.plot(x, x_sin)
```
