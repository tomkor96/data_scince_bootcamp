```python
import matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

Wykres liniowy

```python
plt.plot([0, 1], [0, 3])
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label = 'line', color = 'green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('wykres')
_ = plt.legend()
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
```

Tworzenie wielu wykresów

```python
fig = plt.figure(figsize = (12, 8)) # zmiana rozmiaru wykresu
plt.subplot(131) # 1 - liczba wierszy w siatce wykresów,
# 2 - liczba kolumn w siatce wykresów
# 1 - odniesienie się do wykresu w siatce
plt.plot([0 , 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 1, 2], [3, 1, 4], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [3, 1, 4], 'ro')
```

Wykresy słupkowe

```python
plt.bar(x = [0, 1, 2, 3], height = [1.2, 3.2, 2.2, 4.8])
```

```python
plt.bar(x = ['a', 'b', 'c', 'd'], height = [1.2, 3.2, 2.2, 4.8])
```

```python
plt.barh(y = ['a', 'b', 'c', 'd'], width = [1.2, 3.2, 2.2, 4.8])
# barh - horyzontalny
```

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x = x1, y = x2)
```

```python
x3 = np.random.randn(300) * 50

plt.scatter(x = x1, y = x2, s = x3) # s - size, rozmiar każdej z kropek
```

```python
plt.scatter(x = x1, y = x2, s = x3, alpha = 0.5)
```
